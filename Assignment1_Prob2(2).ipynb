```python
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
img=cv2.imread('test.png',-1)
cv2_imshow(img)
print(img)
```

```python
img.shape
```

```python
p=int(img.shape[0]/2)
print(p)
```

```python
q=int(img.shape[1]/2)
print(q)
```

```python
for i in range(-25+ p):
  for j in range(-25+q):
    img[i][j]=[0,0,0]
```

```python
new_img=img
from matplotlib import pyplot as plt
plt.xticks([]),plt.yticks([])
plt.imshow(new_img)
plt.savefig('output2.png')
```
