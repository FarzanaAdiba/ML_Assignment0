```python
from google.colab.patches import cv2_imshow
import numpy as np
#import cv2
img=cv2.imread('test.png',-1)
cv2_imshow(img)
print(img)
```

```python
r=np.array([255,0,0])
g=np.array([0,255,0])
b=np.array([0,0,255])
```

```python
for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    dist1=np.linalg.norm(img[i][j]-r)
    dist2=np.linalg.norm(img[i][j]-g)
    dist3=np.linalg.norm(img[i][j]-b)
dist=[dist1,dist2,dist3]
img[i][j]=min(dist)
```

```python
from matplotlib import pyplot as plt
plt.xticks([]),plt.yticks([])
plt.imshow(img)
plt.savefig('output1.png')
```
